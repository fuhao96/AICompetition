```python
import cv2
from opcua import Client
from opcua import ua
import datetime
import os
from ultralytics import YOLO

model = YOLO("best_substack.pt")

# Create a list of folder names
folders = ['pass', 'fail']

# Loop through the list of folder names
for folder in folders:
    # Check if the folder already exists
    if not os.path.exists(folder):
        # If it doesn't exist, create the folder
        os.makedirs(folder)

url = ""   #DR3-OP10-02
client = Client(url)
client.set_user("")
client.set_password("")
client.connect()

Ready_node = client.get_node("ns=3;s=\"ST1500_C947_Camera_TRACK_DROP_SUB-STACK\".\"Ready\"")
Alarm_node = client.get_node("ns=3;s=\"ST1500_C947_Camera_TRACK_DROP_SUB-STACK\".\"Alarm\"")

varT = ua.DataValue(ua.Variant(True, ua.VariantType.Boolean))
varF = ua.DataValue(ua.Variant(False, ua.VariantType.Boolean))

# # define a video capture object
# # vid = cv2.VideoCapture("rtsp://gfshap01:Zd4Wj7Es@10.121.121.21/profile2/media.smp")
vid = cv2.VideoCapture("")
# vid =  cv2.VideoCapture(0)

counter = 0

t1 = datetime.datetime.now()

time_gap = 20

cv2.namedWindow("realtime",0);
cv2.namedWindow("result",0);


while(True):
    ret, frame = vid.read()
    
    t2 = datetime.datetime.now()
    t3 = t2 - t1
    if (t3.seconds > time_gap) and Ready_node.get_value():
    # if (t3.seconds > time_gap):
        res = model.predict(frame,conf = 0.7)
        res_plotted = res[0].plot()
        cv2.resizeWindow("result", 800, 600);
        cv2.imshow("result", res_plotted)
        cv2.waitKey(1)
        print(t2)
        filename = str(t2)
        f1 = filename[0:19].replace(' ','_')
        f2 = f1.replace(':','_')
        if int(str(res[0].boxes.shape)[12]) > 0:
            Alarm_node.set_value(varT)
            print("varT")
            cv2.imwrite('fail/'+f2+'.jpg', res_plotted)
        else:
            Alarm_node.set_value(varF)
            print("varF")
            cv2.imwrite('pass/'+f2+'.jpg', res_plotted)
        t1 = datetime.datetime.now()

    # res = model(frame)
    cv2.resizeWindow("realtime", 800, 600);
    cv2.imshow("realtime", frame)
    # cv2.imshow("result", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
  
# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
client.disconnect()
```
