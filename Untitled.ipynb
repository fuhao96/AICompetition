```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('train.csv', sep='\t')

# 替换标签为数字
label_dict = {
    '文学': 1,
    '童书': 2,
    '工业技术': 3,
    '大中专教材教辅': 4,
    '中小学教辅': 5,
    '艺术': 6,
    '管理': 7,
    '社会科学': 8,
    '建筑': 9,
    '小说': 10,
    '计算机与互联网': 11,
    '历史': 12,
    '法律': 13,
    '医学': 14,
    '科学与自然': 15,
    '外语学习': 16,
    '政治/军事': 17,
    '考试': 18,
    '哲学/宗教': 19,
    '励志与成功': 20,
    '经济': 21,
    '文化': 22,
    '青春文学': 23,
    '传记': 24,
    '农业/林业': 25,
    '动漫': 26,
    '烹饪/美食': 27,
    '旅游/地图': 28,
    '育儿/家教': 29,
    '科普读物': 30,
    '国学/古籍': 31,
    '孕产/胎教': 32,
    '健身与保健': 33,
    '金融与投资': 34,
    '婚恋与两性': 35
}

df['label'] = df['label'].map(label_dict)

# 删除包含NaN值的行
df.dropna(subset=['label'], inplace=True)

# 按标签排序
df.sort_values('label', inplace=True)

# 生成新的txt文件
with open('train.txt', 'w', encoding='utf-8') as f:
    for index, row in df.iterrows():
        label = str(int(row['label']))
        title = row['title']
        desc = row['desc']
        f.write(title + desc + '\t' + label + '\n')
```

```python
# 读取原始txt文件
with open('dev.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# 按照标签排序
lines.sort(key=lambda x: x.strip().split('\t')[1])

# 反转描述和标签的顺序，并生成新的txt文件
with open('dev2.txt', 'w', encoding='utf-8') as f:
    for line in lines:
        parts = line.strip().split('\t')
        if len(parts) == 2:
            desc = parts[0]
            label = parts[1]
            f.write(label + '\t' + desc + '\n')
        else:
            # 如果原始txt文件的行不符合预期的格式，直接写入原始行
            f.write(line)
```

```python
import os

def check_and_clean_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    cleaned_lines = []
    for line in lines:
        line = line.strip()
        if not line:
            continue

        parts = line.split('\t')
        if len(parts) != 2:
            continue

        content, label = parts
        try:
            label = int(label)
            cleaned_lines.append(line)
        except ValueError:
            continue

    with open(output_file, 'w', encoding='utf-8') as file:
        file.write('\n'.join(cleaned_lines))

    print("File cleaned and saved successfully.")

# 指定输入文件和输出文件的路径
input_file = 'dev.txt'
output_file = 'dev.txt'

# 检查并清理文件
check_and_clean_file(input_file, output_file)
```

```python
import pandas as pd

# 读取CSV文件
data = pd.read_csv('test.csv', sep='\t')

# 筛选label符合条件的数据
valid_labels = [
    '文学', '童书', '工业技术', '大中专教材教辅', '中小学教辅',
    '艺术', '管理', '社会科学', '建筑', '小说',
    '计算机与互联网', '历史', '法律', '医学', '科学与自然',
    '外语学习', '政治/军事', '考试', '哲学/宗教', '励志与成功',
    '经济', '文化', '青春文学', '传记', '农业/林业',
    '动漫', '烹饪/美食', '旅游/地图', '育儿/家教', '科普读物',
    '国学/古籍', '孕产/胎教', '健身与保健', '金融与投资', '婚恋与两性'
]

filtered_data = data[data['label'].isin(valid_labels)]

# 按照label的顺序排序数据
filtered_data['label'] = pd.Categorical(filtered_data['label'], categories=valid_labels, ordered=True)
sorted_data = filtered_data.sort_values('label')

# 生成新的UTF-8文本文件
output_text = sorted_data['title'].astype(str) + sorted_data['desc'].astype(str) + '\t' + sorted_data['label'].astype(str)
output_text.to_csv('test4.txt', index=False, header=False, encoding='utf-8')
```

```python
def check_and_clean_file(file_path):
    lines_to_keep = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            data = line.split('\t')
            if len(data) == 2:
                lines_to_keep.append(line)
            else:
                print(f"Ignoring invalid line: {line}")
    
    # 保存修改后的文件
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write('\n'.join(lines_to_keep))

# 指定要检查的文件路径
file_path = 'dev.txt'

# 调用函数进行检查和清理
check_and_clean_file(file_path)
```
